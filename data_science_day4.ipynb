### Decision tree Classification

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
df
```

## separate input and output variable

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

```python
y
```

```python
y.value_counts()
```

### Visualize the data





```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c = y)
```

### Import the Class

```python
from sklearn.tree import DecisionTreeClassifier
```

## Create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

### Train the Model

```python
classifier.fit(x, y)
```

### Plot the Tree 

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize= (12, 20))
plot_tree(classifier, fontsize=7, feature_names=['age', 'sal'], 
          class_names=['y', 'n'], filled = True, rounded= True);
```

#### Predict on new data

```python
new = pd.DataFrame({
    'Age' : [ 22, 45, 54,67],
    'EstimatedSalary' : [45000, 89000, 23000, 134000]
})
```

```python
new
```

```python
classifier.predict(new)
```

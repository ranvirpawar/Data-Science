```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

### import the class

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

### Train the model


```python
reg.fit(x, y)
```

### predict on unseen data

```python
disp = 221
hp = 102
wt = 2.91
reg.predict([[disp, hp, wt]])
```
